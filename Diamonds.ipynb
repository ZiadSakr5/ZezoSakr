```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/diamonds.csv')
df
```

```python
df.info()
```

```python
df['z'] = pd.to_numeric(df['z'], errors='coerce')
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df.describe()
```

```python
df = df[df['price'] >= 0]
df.describe()
```

```python
#before dropping nulls
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
#after dropping nulls
df.shape
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    plt.figure()
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

**Most diamonds are between 0.1 and 2 carats. 55 and 65 depth and cheaper**

```python
plt.figure()
sns.scatterplot(data=df, x='carat', y='depth')
plt.title(f"{num_cols[0]} vs {num_cols[1]}")
plt.show()
```

**outlairs are diamonds that have very low or very high depth , and high carat**

```python
x_col, y_col, z_col = num_cols[:3]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df[x_col],
    df[y_col],
    df[z_col],
    alpha=0.6,
    s=40
)

ax.set_xlabel(x_col)
ax.set_ylabel(y_col)
ax.set_zlabel(z_col)
plt.title(f"3D Scatter: {x_col} vs {y_col} vs {z_col}")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, RobustScaler, PowerTransformer
```

```python
catigorical_columns = ['cut', 'color', 'clarity']
```

```python
le_df = df.copy()
LabelEncoder = LabelEncoder()
for col in catigorical_columns:
    le_df[col] = LabelEncoder.fit_transform(le_df[col])
```

```python
ohe_df = df.copy()
ohe = OneHotEncoder(sparse_output=False, drop='first')
ohe_df = pd.DataFrame(ohe.fit_transform(ohe_df[catigorical_columns]))
ohe_df.columns = ohe.get_feature_names_out(catigorical_columns)
ohe_df = pd.concat([ohe_df, df.drop(columns=catigorical_columns)], axis=1)
```

```python
print(ohe_df.head())
```

```python
scaler = StandardScaler()
scaled_array = scaler.fit_transform(df[num_cols])
scaled_df = pd.DataFrame(scaled_array, columns=num_cols)
print(scaled_df.head())

for col in num_cols:
    plt.figure()
    plt.hist(scaled_df[col], bins=30, density=True)
    plt.title(f"StandardScaler – {col}")
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

**i used standardscaler because it is based on mean value of 0 and standard deviation value of 0**

Why we scale data ?

1-linear/logistic regression, neural networks converge much faster when input data is scaled

2-Improved Model Performance and Stability

3-Better Generalization , Scaling helps reduce overfitting

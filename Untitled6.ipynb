```python
import numpy as np

inputs = np.array([[0.05, 0.10]])
targets = np.array([[0.01, 0.99]])

b1 = 0.5
b2 = 0.7

# 2. Initialize weights randomly from the interval [-0.5, 0.5]
np.random.seed(42)
W_hidden = np.random.uniform(-0.5, 0.5, size=(2, 2))
W_output = np.random.uniform(-0.5, 0.5, size=(2, 2))

learning_rate = 0.5

def tanh_activation(x):
    return np.tanh(x)

def tanh_derivative(output):
    return 1.0 - output**2


print(f"Initial Hidden Weights:\n{W_hidden}\n")
print(f"Initial Output Weights:\n{W_output}\n")

# ==========================================
#              FORWARD PASS
# ==========================================


net_hidden = np.dot(inputs, W_hidden) + b1
out_hidden = tanh_activation(net_hidden)


net_output = np.dot(out_hidden, W_output) + b2
final_output = tanh_activation(net_output)
print("--- Forward Pass Results ---")
print(f"Initial Outputs: {final_output[0]}")

total_error = 0.5 * np.sum((targets - final_output) ** 2)
print(f"Initial Total Error: {total_error:.6f}\n")


# ==========================================
#            BACKPROPAGATION
# ==========================================


error_output = final_output - targets
delta_output = error_output * tanh_derivative(final_output)

error_hidden = np.dot(delta_output, W_output.T)
delta_hidden = error_hidden * tanh_derivative(out_hidden)

# Update Weights and Biases (Gradient Descent)
# New Weight = Old Weight - (Learning Rate * Gradient)
W_output -= learning_rate * np.dot(out_hidden.T, delta_output)
W_hidden -= learning_rate * np.dot(inputs.T, delta_hidden)

# Update biases
b2 -= learning_rate * np.sum(delta_output)
b1 -= learning_rate * np.sum(delta_hidden)

print("--- Weights After One Backpropagation Step ---")
print(f"Updated Hidden Weights:\n{W_hidden}\n")
print(f"Updated Output Weights:\n{W_output}\n")


# ==========================================
#      FORWARD PASS (After Updating)
# ==========================================

new_net_hidden = np.dot(inputs, W_hidden) + b1
new_out_hidden = tanh_activation(new_net_hidden)

new_net_output = np.dot(new_out_hidden, W_output) + b2
new_final_output = tanh_activation(new_net_output)

print("--- Forward Pass Results (After Update) ---")
print(f"New Outputs: {new_final_output[0]}")

new_total_error = 0.5 * np.sum((targets - new_final_output) ** 2)
print(f"New Total Error: {new_total_error:.6f}")
```
